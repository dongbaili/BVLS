```python
import numpy as np
import torch
from algorithms import LogisticRegression
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression(input_dim=X_train.shape[1], output_dim=1, alpha=0.1, epochs=1000, lr=0.0001)
model.fit(X_train, y_train)
model.predict(X_test)
print(accuracy_score(y_test, model.predict(X_test)))
```

```python
import matplotlib.pyplot as plt
erm_weight = [0.367, 0.4700, 0.5563, 0.9151, 1.243]
lasso_weight = [0.1409, 0.2030, 0.5528, 0.7126, 0.9048]
x = [0.5, 0.6, 0.7, 0.8, 0.9]
plt.figure(dpi=240)
plt.plot(x, erm_weight, label='ERM')
plt.plot(x, lasso_weight, label='weight Lasso')
plt.title("Model coefficient on spurious feature")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression as skLogisticRegression
model2 = skLogisticRegression()
model2.fit(X_train, y_train)
yp2 = model2.predict(X_test)
print(accuracy_score(y_test, yp2))
```
